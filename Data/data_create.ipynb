```python
import json
import time
import pandas as pd
from tqdm import tqdm
from openai import OpenAI

# =========================
# 1ï¸âƒ£ Read API Key
# =========================

with open("api_deepseek.txt", "r", encoding="utf-8") as f:
    api_key = f.read().strip()

client = OpenAI(
    api_key=api_key,
    base_url="https://api.deepseek.com"
)


# =========================
# 2ï¸âƒ£ Load human seeds
# =========================
with open("llm_seeds_zh.json", "r", encoding="utf-8") as f:
    seeds = json.load(f)

print(f"âœ… Loaded {len(seeds)} human seeds")

df_seeds = pd.DataFrame(seeds)

# =========================
# 3ï¸âƒ£ Bucket by dimension
# =========================
regions = df_seeds[df_seeds["dimension"] == "region"]
ages = df_seeds[df_seeds["dimension"] == "age"]
educations = df_seeds[df_seeds["dimension"] == "education"]

# =========================
# 4ï¸âƒ£ Strict constraint prompt templates
# =========================
PROMPTS = {
    "region": """
You are generating Chinese social stereotype detection data.

Based on the example sentence below, generate {N} new Chinese sentences that:

1. Strictly belong to the stereotype dimension: REGION ONLY.
2. The sentence must involve a Chinese region or province.
3. Preserve the same type and strength of stereotype implication.
4. Do NOT introduce any other attributes such as gender, age, education, or profession.
5. Do NOT involve politics, diseases, extreme hate, or illegal content.
6. Use natural daily Chinese.
7. Each sentence must stand alone.
8. Output ONLY the Chinese sentences, one per line, without any explanation or numbering.

Example:
{SEED}
""",

    "age": """
You are generating Chinese social stereotype detection data.

Based on the example sentence below, generate {N} new Chinese sentences that:

1. Strictly belong to the stereotype dimension: AGE ONLY.
2. The sentence must involve an age group (youth, middle-aged, elderly).
3. Preserve the same type and strength of stereotype implication.
4. Do NOT introduce region, gender, education, profession or nationality.
5. Do NOT involve diseases, politics, extreme hate, or illegal content.
6. Use natural daily Chinese.
7. Each sentence must stand alone.
8. Output ONLY the Chinese sentences, one per line, without any explanation or numbering.

Example:
{SEED}
""",

    "education": """
You are generating Chinese social stereotype detection data.

Based on the example sentence below, generate {N} new Chinese sentences that:

1. Strictly belong to the stereotype dimension: EDUCATION ONLY.
2. The sentence must involve educational background or academic level.
3. Preserve the same type and strength of stereotype implication.
4. Do NOT introduce region, age, gender, profession or nationality.
5. Do NOT involve politics, diseases, extreme hate, or illegal content.
6. Use natural daily Chinese.
7. Each sentence must stand alone.
8. Output ONLY the Chinese sentences, one per line, without any explanation or numbering.

Example:
{SEED}
"""
}

# =========================
# 5ï¸âƒ£ Generation function (new API version)
# =========================
def build_multi_seed_prompt_v2(dimension, seeds, n_per_seed):
    seeds_block = "\n".join([
        f"[Example {i+1}] {s}" for i, s in enumerate(seeds)
    ])

    return f"""
You are generating Chinese social stereotype detection data.

For EACH of the following example sentences, generate EXACTLY {n_per_seed} new Chinese sentences.

Strict rules:
1. Each generated sentence must strictly belong to the stereotype dimension: {dimension.upper()}.
2. Each generated sentence must preserve the SAME type and strength of stereotype implication as its corresponding example.
3. Do NOT introduce any new attributes (gender, region, age, education, profession, nationality).
4. Do NOT involve politics, diseases, extreme hate, or illegal content.
5. Use natural daily Chinese.
6. Output format MUST follow this template EXACTLY:

[Example 1]
1-1. ...
1-2. ...
...
1-{n_per_seed}. ...

[Example 2]
2-1. ...
2-2. ...
...

(continue until all examples are completed)

Examples:
{seeds_block}
""".strip()

def safe_llm_generate_deepseek(prompt):
    max_retry = 6
    wait_time = 5  

    for attempt in range(max_retry):
        try:
            response = client.chat.completions.create(
                model="deepseek-chat",
                messages=[{"role": "user", "content": prompt}],
                temperature=0.7
            )
            return response.choices[0].message.content.strip()

        except Exception as e:
            print(f"âš ï¸ DeepSeek API Error, retry {attempt+1}/{max_retry}: {e}")
            time.sleep(wait_time)
            wait_time *= 2 # Exponential Backoff (20s â†’ 40s â†’ 80s)

    return ""

# =========================
# 6ï¸âƒ£ Batch generation main logic
# =========================
def generate_for_dimension_batch_v2(df, dimension, n_per_seed, output_file, batch_size=5):
    results = []

    texts = df["text_zh"].tolist()
    labels = df["label"].tolist()

    total_batches = (len(texts) + batch_size - 1) // batch_size

    for i in range(total_batches):
        batch_seeds = texts[i * batch_size:(i + 1) * batch_size]
        batch_labels = labels[i * batch_size:(i + 1) * batch_size]

        prompt = build_multi_seed_prompt_v2(dimension, batch_seeds, n_per_seed)
        output = safe_llm_generate_deepseek(prompt)

        if output == "":
            continue

        lines = [x.strip() for x in output.split("\n") if "-" in x]

        expected = len(batch_seeds) * n_per_seed

        if len(lines) < expected:
            print(f"âš ï¸ Output still shorter than expected: {len(lines)}/{expected}")
            continue  # è¿™ä¸€æ‰¹ç›´æ¥ä½œåºŸï¼Œä¸æ±¡æŸ“æ•°æ®

        idx = 0
        for seed_label in batch_labels:
            for _ in range(n_per_seed):
                results.append({
                    "dimension": dimension,
                    "text_zh": lines[idx].split(".", 1)[-1].strip(),
                    "label": seed_label,
                    "source": "llm_generated"
                })
                idx += 1

        # âœ… å®æ—¶è½ç›˜ï¼ˆæ–­ç‚¹å‹å¥½ï¼‰
        pd.DataFrame(results).to_csv(output_file, index=False, encoding="utf-8-sig")

        print(f"âœ… {dimension} batch {i+1}/{total_batches} saved ({len(results)})")

        time.sleep(5)  # âœ… DeepSeek ä¸éœ€è¦ 20 ç§’

    print(f"ğŸ‰ {dimension} generation fully completed!")

```

```python
# =========================
# 7ï¸âƒ£ Start batch generation for three dimensions (2~3x oversampling)
# =========================
generate_for_dimension_batch_v2(regions, "region", n_per_seed=8, output_file="generated_region.csv")
generate_for_dimension_batch_v2(ages, "age", n_per_seed=8, output_file="generated_age.csv")
generate_for_dimension_batch_v2(educations, "education", n_per_seed=8, output_file="generated_education.csv")

print("\nğŸ‰ Step 3 batch augmentation completed!")
```
